```
df=pd.read_csv("/content/drive/My Drive/DM_Project/stock_data_with_ma.csv")
df
```

```
RSIFinal=[]
company=df['comp'][0]
company
index={}
start=0
for i in range(len(df)):
  if company!=df['comp'][i]:
    index[company]=[start,i-1]
    company=df['comp'][i]
    start=i
index[company]=[start,i]
```

```
index
# index['GAIL'][1]
```

```
RSICSV=[]
for key,value in index.items():
  
  change=[]
  upwardChange=[]
  downwardChange=[]
  avgUpwardMovement=[]
  avgDownwardMovement=[]
  relativeStrength=[]
  RSI=[]
  start=value[0]
  end=value[1]
  
  change.append(0)
  for i in range(start+1,end+1):
    change.append(df['adj_close'][i]-df['adj_close'][i-1])
  # print(len(change))
  for i in range(0,len(change)):
    if(change[i]>=0):
      upwardChange.append(change[i])
      downwardChange.append(0)
    else:
      downwardChange.append(abs(change[i]))
      upwardChange.append(0)
  # print(len(upwardChange),len(downwardChange))
  loop=15
  up=0
  down=0
  avgUpwardMovement.append(0)
  avgDownwardMovement.append(0)
  for j in range(1,loop+1):
    up=0
    down=0
    for i in range(1,j):
      up+=upwardChange[i]/j
      down+=downwardChange[i]/j
    # up/=loop
    # down/=loop
    avgUpwardMovement.append(up)
    avgDownwardMovement.append(down) 
  # print(avgUpwardMovement)
  # cnt=loop
  for i in range(loop+1,len(upwardChange)):
    avgUpwardMovement.append((avgUpwardMovement[i-1]*(loop-1)+upwardChange[i])/loop)
    avgDownwardMovement.append((avgDownwardMovement[i-1]*(loop-1)+downwardChange[i])/loop)
    # cnt+=1
  # print(len(avgDownwardMovement), len(avgDownwardMovement))
  relativeStrength.append(0)
  for i in range(1,len(avgUpwardMovement)):
    # print(i)
    if(avgDownwardMovement[i]!=0):
      relativeStrength.append(avgUpwardMovement[i]/avgDownwardMovement[i])
    else:
      relativeStrength.append(0)
  # print(len(relativeStrength))

  for i in range(len(relativeStrength)):
    RSI.append(100-(100/(relativeStrength[i]+1)))
  print(len(RSI))

  for i in range(len(RSI)):
    # if(i<15):
    #   RSICSV.append('')
    # else:
      RSICSV.append(RSI[i])
  # print(len(RSICSV))
  # RSIFinal.append(RSI)
  # print(len(RSICSV))

  
```
