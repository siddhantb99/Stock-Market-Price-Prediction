```
import ta
```

```
def wr(high, low, close, lbp=15, fillna=False):
  return ta.momentum.WilliamsRIndicator(high=high, low=low, close=close, lbp=lbp, fillna=fillna).wr()
wr_temp=wr(high=df['high'],low=df['low'],close=df['close'])
```

```
df['wr']=wr_temp
df.to_csv('stock_data_rsi.csv')
```

```
df
```

```
def stoch(high, low, close, n=14, d_n=3, fillna=False):
  return ta.momentum.StochasticOscillator(high=high, low=low, close=close, n=n, d_n=d_n, fillna=fillna).stoch()
```

```
stoch_temp=stoch(high=df['high'],low=df['low'],close=df['close'])
```

```
df['stoch']=stoch_temp
```
